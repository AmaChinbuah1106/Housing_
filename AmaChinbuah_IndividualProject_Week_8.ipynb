In the peer programming project, we created a data frame representing our store’s products and sales.
In this exercise, we’re going to extend that data frame, quite literally.

The backstory for this exercise is as follows: Our local government is thinking about imposing a
sales tax, and is thinking about 15, 20, and 25 percent rates. Show how much less you would net
with each of these tax amounts by adding columns to the data frame for current income, as well.

### Importing Numpy and Pandas Libraries

```python
import pandas as pd
import numpy as np
```

### Creating a DataFrame from the dictionary 

```python
df=pd.DataFrame([{'product_id':23, 'name':'computer', 'wholesale_price': 500, 'retail_price':1000, 'sales':100}, 
{'product_id':96, 'name':'Python Workout', 'wholesale_price': 35,'retail_price':75, 'sales':1000},
{'product_id':97, 'name':'Pandas Workout', 'wholesale_price': 35, 'retail_price':75, 'sales':500},
{'product_id':15, 'name':'banana', 'wholesale_price': 0.5,'retail_price':1, 'sales':200},
{'product_id':87, 'name':'sandwich', 'wholesale_price': 3,'retail_price':5, 'sales':300}])
df
```

### Setting the product id as index

```python
df.set_index('product_id', inplace=True)
df
```

### Calulating sales tax of 15, 20, and 25 percent rates on current sales 

```python
df['15%_Sales_Tax']=df['sales']+(df['sales'] * 0.15)
df['20%_Sales_Tax']=df['sales']+(df['sales'] * 0.20)
df['25%_Sales_Tax']=df['sales']+(df['sales'] * 0.25)
df
```

### Calculating the new revenue on products at 15 percent Sales tax 

```python
df['net_revenue@15%_Sales_Tax'] = (df['retail_price'] - df['wholesale_price']) * df['15%_Sales_Tax']
df
```

```python
df[['name','wholesale_price','retail_price','15%_Sales_Tax','net_revenue@15%_Sales_Tax']]
```

### Calculating the new revenue on products at 20 percent Sales tax 

```python
df['net_revenue@20%_Sales_Tax'] = (df['retail_price'] - df['wholesale_price']) * df['20%_Sales_Tax']
df
```

```python
df[['name','wholesale_price','retail_price','20%_Sales_Tax','net_revenue@20%_Sales_Tax']]
```

### Calculating the new revenue on products at 25 percent Sales tax 

```python
df['net_revenue@25%_Sales_Tax'] = (df['retail_price'] - df['wholesale_price']) * df['25%_Sales_Tax']
df
```

```python
df[['name','wholesale_price','retail_price','25%_Sales_Tax','net_revenue@25%_Sales_Tax']]
```
