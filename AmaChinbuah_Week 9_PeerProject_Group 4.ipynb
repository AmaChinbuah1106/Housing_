## Requirement:

### Using the same nyc-ticket-violation dataset, let’s assume that tickets can be dismissed if the license plate, state, and/or street name are all there but without requiring the make of car. Remove rows that are missing one or more of these. How many rows remain?Assuming 100 / ticket, how much money would the city lose as a result of this?

### Step 1 :  Import libraries to help code the project

```python
#importing relevant libraries
import pandas as pd 
import numpy as np
```

### Step 2:  Read and Load data into DataFrame

```python
# nyc_df =pd.read_csv("Python Data Files/nyc-parking-violations-2020.csv", delimiter=',', low_memory=False, on_bad_lines=None, usecols=["Plate ID","Registration State","Street Name"])
# nyc_df.head(5)
```

```python
mylist=[]
for data in pd.read_csv('Python Data Files/nyc-parking-violations-2020.csv', delimiter=',' ,chunksize=20000, low_memory=False):
    mylist.append(data)
nyc_df=pd.concat(mylist,axis=0)
del mylist
```

```python
nyc_df
```

### Step 3: Check for the infomation of the loaded data

```python
nyc_df.info() 

#NB nyc_df.shape can also be used, however, this shows the number of rows and columns only.
```

### Step 4:  Select columns required 

```python
nyc_df=nyc_df[["Plate ID","Registration State","Street Name" ]] #slicing was used and reassigned to the initial DataFrame of nyc_df
nyc_df
```

### Step 5:  Count rows with missing data

```python
nyc_df.isnull().sum()
```

### Step 6:  Drop or Delete rows with missing data 

```python
rem_nyc_df =nyc_df.dropna()
rem_nyc_df.info()
```

### Step 7 : Calculate the number of dropped rows 

```python
dropped_rows=nyc_df.shape[0] - rem_nyc_df.shape[0]
dropped_rows
```

### Step 8 : Print the total number of remaining rows after deleting missing rows 

```python
print(f"The number of rows remaining is {rem_nyc_df.shape[0]} ") # the [0] is being used to slice the row only from the tuple of rows and columns
```

### Step 9 : Finding the 'opportunity' cost of removing data with missing values

```python
print (f"The money loss by the city due to missing data is ${dropped_rows*100}.00 ")
```
